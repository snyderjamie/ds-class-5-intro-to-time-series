# External Data Sources

The package [Pandas DataReader](https://pydata.github.io/pandas-datareader/stable/) lets us pull financial data directly from the web. There's not a lot to add about it. Just read [the docs](https://pydata.github.io/pandas-datareader/stable/) if you want to know more about it.

```python
import pandas as pd
import numpy as np
```

```python
import pandas_datareader.data as web
```

```python
start = pd.Timestamp('2017-01-01')
end = pd.Timestamp('2017-02-01')
```

**S&P 500**

```python
web.DataReader('^SPX', 'stooq', start, end)
```

**Stocks**

```python
web.DataReader(('MSFT', 'AAPL', 'GOOG', 'AMZN', 'FB'), 'morningstar', start, end)
```
